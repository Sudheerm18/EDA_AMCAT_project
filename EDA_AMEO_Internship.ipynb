### What is the Relationship between Salary and 12percentage percentage?

```python
import scipy.stats as stats

def pearson_test(data1, data2, significance_level):
    correlation_coefficient, p_value = stats.pearsonr(data1, data2)
    print('correlation coefficient=%.3f, p-value=%.3f' % (correlation_coefficient, p_value))

    if p_value < significance_level:
        print("Reject null hypothesis: There is significant correlation.")
    else:
        print("Fail to reject null hypothesis: There is no significant correlation.")
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,5),)
sns.scatterplot(data=df,x='12percentage',y='Salary',ax=axs[0])
axs[0].set_title('Scatterplot salary vs 12percentage')
sns.lineplot(data=df,x='12percentage',y='Salary',ax=axs[1])
axs[1].set_title('Scatterplot salary vs 12percentage')

plt.show()
```

```python
pearson_test(df['12percentage'], df['Salary'], significance_level=0.05)
```

+ Employer Consideration: Employers may view a strong performance in 12th grade as indicative of a candidate's academic diligence and potential contribution to the organization, influencing Salary decisions.

## Gender v/s Specialization

```python
df.columns
```

```python
pd.crosstab(index=df['Gender'], columns=df['Specialization Category'])
```

The p-value of the statistic is greater than the significance level, indicating that the variables are independent. Hence, there's no association between gender and specialization.


```python
from scipy.stats import chi2_contingency

gen_spec_df = df[df['Specialization'] != 'Other']
cont_tab = pd.crosstab(index=gen_spec_df['Gender'], columns=gen_spec_df['Specialization Category'])
chi2 = chi2_contingency(cont_tab, correction=False)
print("statistic:", chi2.statistic)
print("p-value:", chi2.pvalue)
```

The p-value of the statistic is greater than the significance level, indicating that the variables are independent. Hence, there's no association between gender and specialization. Surprisingly, there is a significant difference between men's and women's incomes. Women earn slightly less than men.
